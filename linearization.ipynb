# Dynamical Systems: Linearization, Equilibrium Points & Stability

Consider the following general nonlinear system:
$$\dot q_1 = aq_1 - bq_1q_2$$
$$\dot q_2 = bq_1q_2-cq_2$$

where $q_1, q_2 \geq 0$ and $a, b, c \geq 0$ are positive constants.

This can be represented as:

$$\frac{dx}{dt} = f(x)$$

where

$$x = 
\begin{bmatrix} 
q_1 \\ 
q_2 
\end{bmatrix}$$

$$\frac{dx}{dt} = 
\begin{bmatrix}
\dot q_1 \\
\dot q_2 
\end{bmatrix}$$

$$f(x) = 
\begin{bmatrix} 
\dot q_1 \\ 
\dot q_2 
\end{bmatrix}
= 
\begin{bmatrix}
aq_1 - bq_1q_2 \\ 
bq_1q_2-cq_2 
\end{bmatrix}$$

Equilibrium points occur when our system does not change with time. This occurs when:
$$\dot q_1 = 0$$
$$\dot q_2 = 0$$

Therefore, equilibrium points occur when:
$$q_1 (a-bq_2) = 0$$
$$q_2 (bq_1-c) = 0$$

We can solve and show that there are two equilbrium points:

> Equilibrium Point 1: 
$$q_1=0 \text{ , } q_2=0 $$

> Equilibrium Point 2: 
$$q_1= \frac{c}{b} \text{ , } q_2= \frac{a}{b}$$

### Jacobian Linearization Around An Equilibrium Point:
The Jacobian linearization of our nonlinear system is:
$$\frac{dx}{dt} = Ax$$

where $A$ is the Jacobian of our matrix $f(x)$.

Let $\mathbf{f_1} = \mathbf{\dot q_1}$,
$\mathbf{f_2} = \mathbf{\dot q_2}$, and 
$\mathbf{x} = \begin{bmatrix}
\mathbf{q_1} \\
\mathbf{q_2}
\end{bmatrix}$

$$\mathbf{A}
=
\mathbf{\frac{\partial f}{\partial x}}
=
\begin{bmatrix}
\mathbf{\frac{\partial f_1}{\partial q_1}} &
\mathbf{\frac{\partial f_1}{\partial q_2}} \\
\\
\mathbf{\frac{\partial f_2}{\partial q_1}} &
\mathbf{\frac{\partial f_2}{\partial q_2}}
\end{bmatrix}$$

Evaluated at Equilibrium Point 1:
>$$\mathbf{A}
=
\begin{bmatrix}
a-bq_2 &
-bq_1 \\
bq_2 &
bq_1-c
\end{bmatrix}
\Biggr|_{q_1=0 \text{ , } q_2=0}$$
$$\mathbf{A}
=
\begin{bmatrix}
a &
0 \\
0 &
-c
\end{bmatrix}$$
$$det(\mathbf{A}-\lambda I) = 0$$
$$\lambda_1 = a { , }$$
$$\lambda_2 = -c$$
$$\text{Since }\lambda_1 \text{is positive, the system is unstable around this equilibrium point.}$$

Evaluated at Equilibrium Point 2:
>$$\mathbf{A}
=
\begin{bmatrix}
a-bq_2 &
-bq_1 \\
bq_2 &
bq_1-c
\end{bmatrix}
\Biggr|_{q_1=\frac{c}{b} \text{ , } q_2=\frac{a}{b}}$$
$$\mathbf{A}
=
\begin{bmatrix}
0 &
-c \\
a &
0
\end{bmatrix}$$
$$det(\mathbf{A}-\lambda I) = 0$$
$$\lambda_1 = +i\sqrt{ac} { , }$$
$$\lambda_2 = -i\sqrt{ac}$$
$$\text{Since } \lambda_1 \text{ and } \lambda_2 
\text{are complex conjugates, the system is marginally stable around this equilibrium point.}$$

### Simulation of this system:

Choosing arbitrary constants $a, b, c$, we can simulate the response of this system for varying initial conditions. We choose a small time step, $\Delta t$, and assume that $f$ is constant during the interval. Under these assumptions, $q$ can be iteratively updated by
$$q_{t + \Delta T} = q_t + \Delta T \dot{q_t} = q_t + \Delta T f(q)$$

```python
import numpy as np
import matplotlib.pyplot as plt

a = b = c = 1
dt = 0.1

time = np.arange(0, 20, dt)
q_data = np.zeros((len(time), 2))


def dqdt(q):
    return np.array((
        (a*q[0])-(b*q[0]*q[1]),
        (b*q[0]*q[1])-(c*q[1])))

# Initial Conditions:
q = np.array((1, 3))

for idx, t in enumerate(time):
    q = q + dt * dqdt(q)
    q_data[idx,:] = q
    
plt.plot(q_data[:,0], q_data[:,1])
plt.legend(('q1','q2'))
```

```python
# Initial Conditions:
q = np.array((3, 3))

for idx, t in enumerate(time):
    q = q + dt * dqdt(q)
    q_data[idx,:] = q
    
plt.plot(q_data[:,0], q_data[:,1])
plt.legend(('q1','q2'))
```
